```python
import seaborn as sb
import matplotlib.pyplot as plt
import random
```

```python
class Dice:
    def __init__(self, faces: int = 6) -> None:
        self.faces = faces

    def __repr__(self) -> str:
        return f"Dice({self.faces})"

    def roll(self) -> int:
        return random.randint(1, self.faces)
```

```python
dice = Dice()
attempts = 10000
freq = {}

for _ in range(attempts):
    val = dice.roll()
    if val in freq:
        freq[val] += 1
    else:
        freq[val] = 1

ax = sb.barplot(freq)
plt.bar_label(ax.containers[0], fmt=lambda x: f"{x:1.0f}", label_type="center", color="white")
plt.title(f"Dice Rolls - Total Attempts: {attempts}")
plt.xlabel("Value")
plt.ylabel("Frequancy")

plt.show()

for key, val in freq.items():
    freq[key] = val/attempts*100

ax = sb.barplot(freq)
plt.bar_label(ax.containers[0], fmt=lambda x: f"{x:1.2f} %", label_type="center", color="white")
plt.title(f"Dice Rolls - Total Attempts: {attempts}")
plt.xlabel("Value")
plt.ylabel("Frequancy (%)")

plt.show()
```
