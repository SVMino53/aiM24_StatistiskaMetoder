```python
import math
import seaborn as sns
import matplotlib.pyplot as plt
import random
```

# a)

```python
x = {1: 0.05, 2: 0.1, 3: 0.15, 4: 0.05, 5: 0.05,
     6: 0.05, 7: 0.05, 8: 0.05, 9: 0.4, 10: 0.05}

sns.barplot(x)
plt.title("Sannolikhetsfördelning för X\nP(X=x)")
plt.xlabel("x")
plt.ylabel("Sannolikhet")
plt.grid(visible=True, axis="y")

plt.show()
```

# b)

```python
def f_func(data: dict, x: int) -> float:
    keys = data.keys()
    start = min(keys)
    s = 0
    for k in range(start, x + 1):
        s += data[k]
    return s


fx = {k: f_func(x, k) for k in x.keys()}

sns.barplot(fx)
plt.title("Kummulativ Fördelning för X\nF(x)")
plt.xlabel("x")
plt.ylabel("Sannolikhet")
plt.grid(visible=True, axis="y")

plt.show()
```

# c)

```python
e = sum([k*p for k, p in x.items()])
print(f"E(X) = {e}")

v = sum([p*(k - e)**2 for k, p in x.items()])
print(f"V(X) = {v}")

s = math.sqrt(v)
print(f"S(X) = {s}")
```

# d)

## 1-3.

```python
def sample_mean(data: dict, n: int) -> float:
    tests = random.choices(list(data.keys()), data.values(), k=n)
    return sum(tests)/n

stickprov = [sample_mean(x, 1) for _ in range(1000)]

sns.histplot(stickprov, bins=10, stat="probability", kde=True)
plt.xlim(1, 10)
plt.title("Sannolikhetsfördelning för X̄\nn = 1")
plt.xlabel("x\u0304")
plt.ylabel("Sannolikhet")
plt.grid(visible=True, axis="y")

plt.show()
```

## 4.

```python
n_vals = [2, 4, 8, 12, 15, 20, 24, 30, 35, 50, 100, 200]

for n in n_vals:
    stickprov = [sample_mean(x, n) for _ in range(1000)]

    sns.histplot(stickprov, bins=10, stat="probability", kde=True)
    plt.xlim(1, 10)
    plt.title(f"Sannolikhetsfördelning för X̄\nn = {n}")
    plt.xlabel("x\u0304")
    plt.ylabel("Sannolikhet")
    plt.grid(visible=True, axis="y")

    plt.show()
```

# e)

Då $n = 1$ får vi en plott som liknar sannolikhetsfördelningen för initiella slumpvariabelln $X$.

När $n$ växer blir plotten mer och mer lik en normalfördelning där medelvärdet $\mu = E(X)$ och standardavvikelsen blir mindre och mindre.
