```python
import math
import random
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
def korrelationskoefficient(x: list, y: list) -> float:
    x_mean = sum(x)/len(x)
    y_mean = sum(y)/len(y)
    ss_x = sum([(val - x_mean)**2 for val in x])
    ss_y = sum([(val - y_mean)**2 for val in y])
    ss_xy = sum([(x_val - x_mean)*(y_val - y_mean) for x_val, y_val in zip(x, y)])
    return ss_xy/math.sqrt(ss_x*ss_y)
```

```python
x = [i + random.random()*10 for i in range(100)]
y = [i + random.random()*10 for i in range(100)]

r = korrelationskoefficient(x, y)

plt.scatter(x, y)
plt.title(f"Korrelationskoefficient = {r:1.4f}")

plt.show()

x = [32, 35, 39, 40, 47, 53, 55]
y = [3500, 3600, 3900, 4300, 4400, 5000, 5100]

r = korrelationskoefficient(x, y)

plt.scatter(x, y)
plt.title(f"Korrelationskoefficient = {r:1.4f}")
plt.xlabel("Yta (m^2)")
plt.ylabel("Hyra (kr)")

plt.show()
```
