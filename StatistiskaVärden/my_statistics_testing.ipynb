```python
import my_statistics_funcs as sf
import matplotlib.pyplot as plt
import random
import math
```

```python
x = random.choices([i for i in range(-10, 11)], [math.comb(20, k) for k in range(21)], k=10)
random.shuffle(x=x)

print(f"Stickprov: {x}")
print()
print(f"Medelvärde: {sf.mean(x)}")
print(f"Kvartil 1: {sf.q1(x)}")
print(f"Kvartil 2 (median): {sf.median(x)}")
print(f"Kvartil 3: {sf.q3(x)}")
print()
print(f"Varians (stickprov): {sf.variance(x)}")
print(f"Standardavvikelse (stickprov): {sf.standard_dev(x)}")
print(f"Varians (population): {sf.pvariance(x)}")
print(f"Standardavvikelse (population): {sf.pstandard_dev(x)}")
print(f"MAD: {sf.mad(x)}")
```

```python
exp_dist = sf.ExponentialDist(1.5)
norm_dist = sf.NormalDist(0, 1)

fig, ax = plt.subplots()

exp_dist.graph(ax=ax)
exp_dist.fgraph(ax=ax)
ax.set_title("Exponential Distribution")
ax.set_ylabel("Probability")
ax.legend(["f(x)", "P(X <= x)"])

plt.show()

fig, ax = plt.subplots()

norm_dist.graph(ax=ax)
norm_dist.fgraph(ax=ax)
ax.set_title("Normal Distribution")
ax.set_ylabel("Probability")
ax.legend(["f(x)", "P(X <= x)"])

plt.show()
```
