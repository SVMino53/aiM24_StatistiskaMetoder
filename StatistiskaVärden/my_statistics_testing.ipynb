```python
import my_statistics_funcs as sf
import matplotlib.pyplot as plt
import random
import math
```

```python
x = random.choices([i for i in range(-10, 11)], [math.comb(20, k) for k in range(21)], k=10)
random.shuffle(x=x)

print(f"Stickprov: {x}")
print()
print(f"Medelvärde: {sf.mean(x)}")
print(f"Kvartil 1: {sf.q1(x)}")
print(f"Kvartil 2 (median): {sf.median(x)}")
print(f"Kvartil 3: {sf.q3(x)}")
print()
print(f"Varians (stickprov): {sf.variance(x)}")
print(f"Standardavvikelse (stickprov): {sf.standard_dev(x)}")
print(f"Varians (population): {sf.pvariance(x)}")
print(f"Standardavvikelse (population): {sf.pstandard_dev(x)}")
print(f"MAD: {sf.mad(x)}")
```

```python
def show_contdist_graphs(dist: sf.ContinuousDist, title: str = "Continuous Distribution") -> None:
    fig, ax = plt.subplots()

    dist.graph(ax=ax)
    dist.fgraph(ax=ax)
    ax.set_title(title)
    ax.set_ylabel("Probability")
    ax.legend(["f(x)", "P(X \u2264 x)"])

    plt.show()
```

```python
exp_dist = sf.ExponentialDist(1.5)
norm_dist = sf.NormalDist(100, 0.5)
z_dist = sf.ZDist()
t1_dist = sf.TDist(1)
t2_dist = sf.TDist(2)
t3_dist = sf.TDist(3)
t5_dist = sf.TDist(5)
t10_dist = sf.TDist(10)

show_contdist_graphs(exp_dist, f"Exponential Distribution\nX~Exp({exp_dist.mu})")
show_contdist_graphs(norm_dist, f"Normal Distribution\nX~N({norm_dist.mu}, {norm_dist.s})")
show_contdist_graphs(z_dist, "Standard Normal Distribution\nX~N(0, 1)")
show_contdist_graphs(t1_dist, "Student's T-Distribution\nX~T(1)")
show_contdist_graphs(t2_dist, "Student's T-Distribution\nX~T(2)")
show_contdist_graphs(t3_dist, "Student's T-Distribution\nX~T(3)")
show_contdist_graphs(t5_dist, "Student's T-Distribution\nX~T(5)")
show_contdist_graphs(t10_dist, "Student's T-Distribution\nX~T(10)")
```

```python
fig, ax = plt.subplots()

t1_dist.graph(ax=ax)
t2_dist.graph(ax=ax)
t3_dist.graph(ax=ax)
t5_dist.graph(ax=ax)
t10_dist.graph(ax=ax)
z_dist.graph(ax=ax)
ax.legend(["t1", "t2", "t3", "t5", "t10", "z"])
ax.set_title("T- and Z-Distribution\n(PDF)")
ax.set_ylabel("Probability")

plt.show()

fig, ax = plt.subplots()

t1_dist.fgraph(ax=ax)
t2_dist.fgraph(ax=ax)
t3_dist.fgraph(ax=ax)
t5_dist.fgraph(ax=ax)
t10_dist.fgraph(ax=ax)
z_dist.fgraph(ax=ax)
ax.legend(["t1", "t2", "t3", "t5", "t10", "z"])
ax.set_title("T- and Z-Distribution\n(CDF)")
ax.set_ylabel("Probability")

plt.show()
```
